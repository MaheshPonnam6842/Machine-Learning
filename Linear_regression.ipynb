from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
import seaborn as sns

```python
df=pd.read_csv('home_prices.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
model=LinearRegression()
```

```python
X=df[['area_sqr_ft']]
y=df['price_lakhs']
```

```python
model.fit(X,y)
```

```python
model.predict([[1200],[1600]])
```

Internal mechanism

```python
model.coef_,model.intercept_
```

y=mx+b where b=intercept , m=slope(coef_) y=price(target)  x = independent variable (Area)

```python
price=model.coef_*1200+model.intercept_
```

```python
price_1=model.coef_*1600+model.intercept_
```

```python
price,price_1
```
